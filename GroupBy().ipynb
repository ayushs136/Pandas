# THE .agg() method

```python
import pandas as pd
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors  = fortune.groupby("Sector") # grouping a object
fortune.head(3) 
```

```python
sectors['Employees'].mean()
```

```python
sectors.agg('Employees').sum()
```

```python
sectors.agg({'Revenue': 'sum',
            'Profits':'sum',
             'Employees': 'mean'
            })
```

```python
sectors.agg(['size', 'sum', 'mean'])
```

# iterating through the Groups

```python
import pandas as pd
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors  = fortune.groupby("Sector") # grouping a object
fortune.head(3) 
```

```python
sectors['Profits'].max()
```

```python
df = pd.DataFrame(columns = fortune.columns)
```

```python
df
```

```python
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, 'Revenue')
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
cities = fortune.groupby('Location')
```

```python
df = pd.DataFrame(columns = fortune.columns)
```

```python
df
```

```python
for city, data in cities:
    h = data.nlargest(1, 'Revenue')
    df  = df.append(h)
```

```python
df
```

# pandas-GroupBy

```python

import pandas as pd
import numpy as np
```

```python
data = {'company': ['GOOG', 'GOOG','MSFT', 'MSFT', 'FB', 'FB'],
       'Person': ['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
       'Sales':[200,120,320,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
company = df.groupby('company')
```

```python
company.mean()
```

```python
company.agg({'Sales': 'sum'})
```

```python
company.sum().loc['FB']
```

```python
df.groupby('company').sum().loc['FB']
```

```python
company.describe().transpose()['FB']
```
